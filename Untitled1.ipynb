```python
# Loops - Laços e oterações

for i in range (5):
  print ("Olá mundo!")
```

```python
# Escrever um programa que recebe um número e retorna somente os números pares.

numero = int(input("Digite um número: "))

for i in range(numero + 1):
  if i % 2 == 0:
    print (i)
```

```python
# Escrever um programa que recebe um número e imprime uma contagem regressiva

numero = int(input("Digite um numero"))

for i in range(numero +1):
  print(numero - i)
```

```python
# Escrever a palavra ao contrario

palavra = str (input("Digite uma palavra: "))
palindromo = ""

for i in (palavra):
  palindromo = i + palindromo
print(palindromo)
```

```python
# Escrever um programa que recebe um numero de 1 a 10 imprime a tabuada.

numero = int(input("Digite um numero"))

for i in range(11):
  print(f"{numero} x {i} = {numero * i}")
```
