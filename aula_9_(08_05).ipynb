```python
# 1 - Contagem siples, o usuário digita um número e o programa imprime do zero até o número.

numero = int(input("Digite um número: "))

for i in range(numero + 1):
    print (i)
```

```python
# 2 - Contagem regressiva, o usuário digita um número e o programa imprime até o zero.

numero = int(input("Digite um numero"))

for i in range(numero +1):
  print(numero - i)
```

```python
# 3 - Números pares, o usuário digita um número e o programa imprime os números pares.

numero = int(input("Digite um número: "))

for i in range(numero + 1):
  if i % 2 == 0:
    print (i)
```

```python
# 4 - Números ímpares, o usuário digita um número e o programa imprime os núemros ímpares.

numero = int(input("Digite um número: "))

for i in range(numero + 1):
  if i % 2 == 1:
    print (i)
```

```python
# 5 - Soma, o usuário digita um número e o programa soma os números do zero até o número.

numero = int(input("Digite um numero"))

for i in range(0,7):
  print(f"{numero} + {i} = {numero + i}")
```

```python
# 6 - Tabuada, o usuário digita um número e o programa imprime a tabuada do número.

numero = int(input("Digite um numero"))

for i in range(11):
  print(f"{numero} x {i} = {numero * i}")
```

```python
# 7 - Contar letras, o usuário digita uma palavra e o programa conta as letras da palavra.

# Entrada
palavra = input("Digite uma palavra: ")

# Processamento
letras = len(palavra)

# Saída
print(f"A palavra '{palavra}' tem {letras} letras.")
```

```python
# 8 - Contar vogais, o usuário digita uma palavra e o programa conta as vogais.

# Entrada

palavra = input("Digite uma palavra:")

# Processamento

vogais = "aeiouAEIOU"
contador = 0

for letra in palavra:
  if letra in vogais:
    contador += 1

# Saída

print(f"A palavra '{palavra}' tem {contador} vogais.")
```

```python
# 9 - Maior número digitado, o usuário digita quatro números e o programa imprime o maior número digitado.

# Entrada

primeiro = int(input("Digite o primeiro número: "))
segundo = int(input("Digite o segundo número: "))
terceiro = int(input("Digite o terceiro número: "))
quarto = int(input("Digite o quarto número: "))

# Processamento

maior = primeiro

if segundo > maior:
  maior = segundo
if terceiro > maior:
  maior = terceiro
if quarto > maior:
  maior = quarto

# Saída

print(f"O maior número digitado foi {maior}.")
```

```python
# 10 - O usuário digita um número entre 01 e 100 e o programa mostra quais são divisíveis por 3 e 5.

# Entrada

numero = int(input("Digite um número entre 1 e 100: "))

# Processamento

print(f"Números divisíveis por 3 e 5 entre 1 e {numero}:")

for i in range(1, numero + 1):
  if i % 3 == 0 and i % 5 == 0:
    print(i)
```
