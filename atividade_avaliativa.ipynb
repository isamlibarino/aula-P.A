Escrever um programa que fala se o número é par ou ímpar

Solicite dois números e mostre qual é o maior (ou se são iguais).

```python
# Entrada

numero1= int (input ("Digite um numero"))
numero2= int (input ("Digite outro numero"))

# Processamento

if numero1 > numero2:
    print (f"O numero {numero1} é maior que {numero2}")

elif numero1 < numero2:
    print (f"O numero {numero2} é maior que {numero1}")

else:
    print (f"O numero {numero1} é igual")
```

Peça a nota de um aluno e informe se ele foi aprovado ou reprovado

```python
# Nota maior ou igual a 7 - aprovado
# Nota maior que 5 e menor que 7 - recuperação
# Nota menor que 5 - reprovado


# Entrada

nota1= float (input("Digite a nota do primeiro bimestre"))
nota2= float (input("Digite a nota do segundo bimestre"))
nota3= float (input("Digite a nota do terceiro bimestre"))
nota4= float (input("Digite a nota do quarto bimestre"))

# Processamento

media= (nota1 + nota2 + nota3 + nota4) /4

if media < 5:
   print (f"voce está reprovado, sua media é {media}")

elif media >= 5 and media < 7:
   print (f"voce está de recuperação, sua media é {media}")

else:
    print (f"voce está aprovado, sua media é {media}")
```

Receba a idade e classifique se é criança, adolescente, adulto ou idoso

```python
# Entrada
nome = str (input("Digite o seu nome"))
idade = int (input("Digite a sua idade"))

# Processamento / Saída

if idade < 12:
    print (f"{nome}, voce é uma criança")

elif idade >= 12 and idade < 18:
    print (f"{nome}, voce é um adolescente")

elif idade >= 18 and idade <60:
    print (f"{nome}, voce é um adulto")

else:
    print (f"{nome}, voce é um idoso")
```

Peça um número e diga se ele é positivo, negativo ou zero.

```python
# Entrada
numero = float(input("Digite um numero"))

# Processamento / Saída

if numero > 0:
    print(f"O numero {numero} é positivo")

elif numero < 0:
    print(f"O numero {numero} é negativo")

else:
    print(f"O numero {numero} é zero")
```

Peça dois números e uma operação (+, -, *, /). Use if/elif para calcular o resultado.

Peça um número de 1 a 7 e mostre o dia correspondente (ex: 1 = domingo).

```python
# Entrada
numero_dia = int(input("Digite um numuero de 1 a 7:"))

# Processamento

if numero_dia == 1:
  print("O dia correspondente é domingo")

elif numero_dia == 2:
  print("O dia correspondente é segunda-feira")

elif numero_dia == 3:
  print("O dia correspondente é terça-feira")

elif numero_dia == 4:
  print("O dia correspondente é quarta-feira")

elif numero_dia == 5:
  print("O dia correspondente é quinta-feira")

elif numero_dia == 6:
  print("O dia correspondente é sexta-feira")

elif numero_dia == 7:
  print("O dia correspondente é sabado")

else:
  print("Numero inválido")
```

Peça o valor de uma compra:
≥ 100: 10% de desconto
≥ 50 e < 100: 5% de desconto
< 50: sem desconto


```python
# Entrada
compra= float (input("Digite o valor da sua compra: "))


if compra >= 100:
  desconto = compra - (compra * 0.10)
  print (f"O valor a pagar é: {desconto}")

elif compra >= 50 and compra < 100:
  desconto = compra - (compra * 0.05)
  print (f"O valor a pagar é: {desconto}")

else:
  print (f"Voce nao obteve desconto, valor a pagar: {compra}")
```
