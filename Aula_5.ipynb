# Exercício Celsius para Fahrenheit

```python
# Calcular a temperatura de Celsius para Fahrenheit

# Entrada

celsius = float( input("Digite quantos graus"))

#Processamento

Resultado = celsius * 9/5 + 32

#Saída
print(f"{celsius}°C é igual a {Resultado}°F")
```
