```python
# Entrada de Decisão (if, elf e else)

# Escrever um programa que compara dois números

# Entradas
a = int (input("digite um número inteiro: "))
b = int (input(" Digite um número inteiro:"))


# Processamento / Saída

if a == b:
  print(f"{a} é igual a {b}")

else:
    print(f"{a} é diferente de {b}")
```

```python
# Escrever um programa que verifica se o número digitado é par ou impar


# Entrada

numero = int (input("Digite um número: "))



# Processamento / Saída

if numero % 2 == 0:

  print (f"O número {numero} é par. ")

else:
  print (f"O número {numero} é impar. ")
```

```python
# Escrever um programa que verifica se a idade corresponde a uma criança ou a um adolescente

# Entrada
nome = str (input(" Digite o seu nome: "))
idade = int (input(" Digite sua idade: "))

# Processamento / Saída

if idade < 12 :
  print (f"{nome}, você é uma Criança ")

elif idade < 18 :
  print (f"{nome}, voce é um  Adolecente")

else:
  print (f"{nome}, você é um  Adulto")
```
