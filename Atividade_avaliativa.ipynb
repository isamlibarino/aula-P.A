  Questão 2 -  Soma dos Números

```python
numero = int(input("Digite um número inteiro positivo: "))

soma = 0
if numero > 0:
    for i in range(1, numero + 1):
        soma += i
    print(f"Resultado: {soma}")
else:
    print("digite um número inteiro positivo.")
```

Questão 3 - Solicite ao usuário um número inteiro e exiba sua tabuada
de 1 até 10.

```python
# Solicite ao usuário um número inteiro e exiba sua tabuada de 1 até 10.

numero = int(input("Digite um numero"))

for i in range(11):
  print(f"{numero} x {i} = {numero * i}")
```

  Questão 4 - Solicite ao usuário um número inteiro positivo e calcule seu fatorial utilizando um laço de repetição.

```python
# Solicite ao usuário um número intero positivo e calcule seu fatorial utilizando um laço de repetição

numero = int(input("Digite um número inteiro positivo: "))

fatorial = 1
if numero > 0:
    for i in range(1, numero + 1):
        fatorial *= i
    print(f"O fatorial de {numero} é {fatorial}")
else:
    print("Digite um número inteiro positivo.")
```

Questão 5 - Solicite ao usuário o valor do saque e informe a menor
quantidade possível de notas para realizar o pagamento.

```python
#Solicite ao usuário o valor do saque e informe a menor quantidade possível de notas para realizar o pagamento.

numero = int(input("Digite o valor do saque: "))

notas = [100, 50, 20, 10, 5, 2, 1]

for nota in notas:
    quantidade = numero // nota
    numero %= nota
    if quantidade > 0:
        print(f"{quantidade} nota(s) de R${nota}")
```
