Escrever um programa que fala se é par ou ímpar

```python
# Entrada

numero= int (input("Digite um numero"))

#Processamento

if numero % 2 ==0:

    print (f"O numero {numero} par")

else:
    print (f"O numero {numero} é impar")
```

```python
 # Escrever um programa que verifica as notas dos quatro bimestres e calcula a media do aluno

 # Não permitir que o usuário digite notas menores que zero ou maiores que 10
 # Média menor que 5 - Reprovado
 # Média maior que 5 e menor que 7 - Recuperação
 # Média maior ou igual a 7 - Aprovado


 # Entrada

 nota1= float (input("Digite a nota do primeiro bimestre"))

 if nota1 < 0 or nota1 > 10:
    print ("Nota inválida, digite sua nota de 0 a 10:")
    nota1= float (input("Digite a nota do primeiro bimestre"))


 nota2= float (input("Digite a nota do segundo bimestre"))

 if nota2 < 0 or nota2 > 10:
    print ("Nota inválida, digite sua nota de 0 a 10:")
    nota2= float (input("Digite a nota do segundo bimestre"))


 nota3= float (input("Digite a nota do terceiro bimestre"))

 if nota3 < 0 or nota3 > 10:
    print ("Nota inválida, digite sua nota de 0 a 10:")
    nota3= float (input("Digite a nota do terceiro bimestre"))


 nota4= float (input("Digite a nota do quarto bimestre"))

 if nota4 < 0 or nota4 > 10:
    print ("Nota inválida, digite sua nota de 0 a 10:")
    nota4= float (input("Digite a nota do quarto bimestre"))

 # Processamento

 media= (nota1 + nota2 + nota3 + nota4)/4

if media < 5:
    print (f"voce está reprovado, sua media é : {media}")

elif media <= 7:

    print (f"voce está de recuperação, sua media é : {media}")

else:
    print (f"voce está aprovado, sua media é :{media}")
```

```python
# Escrever um programa que leia o valor da compra e aplique:

# Até R$ 100 - sem desconto
# R$ 101 - a R$ 500 - 10% de desconto
# Acima de R$ 500 - desconto de 20%

# Mostrar valor a pagar depois de aplicado o desconto.

compra= float (input("Digite o valor da sua compra: "))


if compra > 100:
  desconto = compra - (compra * 0.10)
  print (f"O valor a pagar é: {desconto}")

elif compra > 500:
  desconto = compra - (compra * 0.20)
  print (f"O valor a pagar é: {desconto}")

else:
  print (f"Voce não obteve desconto, valor a pagar: {compra}")
```
